```python
# 绘制sigmoid函数
import matplotlib.pyplot as plt
import numpy as np
def sigmoid(z):
    return 1/(1+np.exp(-z))
z = np.linspace(-10,10,30)
phi_z = sigmoid(z)
plt.plot(z,phi_z)
plt.xlabel("z")
plt.ylabel("$simoid(z)$")
plt.show()
```

```python
#绘制标准正态累积分布函数
from scipy.stats import norm
z=np.linspace(-10,10,300)
phi_z = norm.cdf(z)
plt.plot(z, phi_z)
plt.xlabel("z")
plt.ylabel("$\Phi(z)$")
plt.show()
```

```python
# 逻辑回归的代码实现
## 导入所需的包
from numpy import *
import numpy as np

## 生成数据
np.random.seed(0)
X = np.random.randn(300,5)
Beta = np.mat([2,1,3,0.5,4]).T
err = np.random.randn(300)
err = err.reshape(300,1)
y_prob = 1/(1+np.exp(-X*Beta + err))
y = np.where(y_prob>0.5, 1, 0)
```

```python
## 利用梯度下降计算beta
beta0 = np.mat([1,1,1,1,1]).T
alpha = 0.001

def prob(beta):
    return 1/(1+np.exp(-X*beta))  #计算y=1的概率p(y=1|X)

def loss(beta):
    p = prob(beta)
    return -np.sum((1-y).T*np.log(1-p)+y.T*np.log(p)) #计算损失函数
def grad(beta):
    p = prob(beta)
    return X.T*(p-y)

#开始迭代过程 
betas = []
ls = []
beta = beta0

for i in range(500):
    betas.append(np.array(beta).reshape(5,)) #记录执行过程
    l = loss(beta)
    ls.append(l)
    gradient = grad(beta)  #计算梯度    
    beta = beta - alpha * gradient    
    
#迭代500次之后的结果
print(f"beta = {beta}")

plt.plot(betas)
plt.xlabel("iterations")
plt.ylabel("coefs")
plt.show()

plt.plot(ls)
plt.xlabel("iterations")
plt.ylabel("loss")
plt.title("loss")
plt.show()
```
